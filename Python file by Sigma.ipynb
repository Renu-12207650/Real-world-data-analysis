## Probability Distributions


### Normal distribution to unit price

```python
import scipy.stats as stats
# Extract the 'Unit Price' data
unit_price_data = df['Unit price']
# histogram to visualize the distribution
plt.hist(unit_price_data, bins=20, density=True, alpha=0.6, color='b')

mu, std = stats.norm.fit(unit_price_data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
# probability density function
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.show()
```

## Statistical Inference

### t-test using unit price and customer types

```python
import pandas as pd
from scipy import stats
# Separate data into two groups: "Member" and "Normal"
member_df = df[df["Customer type"] == "Member"]["Unit price"]
normal_df = df[df["Customer type"] == "Normal"]["Unit price"]
# two-sample t-test
t_stat, p_value = stats.ttest_ind(member_df, normal_df, equal_var=False)
# Set the significance level (alpha)
alpha = 0.05
# Make a decision and interpret the results
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in the average unit price between 'Member' and 'Normal' customer types.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in the average unit price between 'Member' and 'Normal' customer types.")
```

#### Correlation Test (Pearson's Correlation)

```python
#  Pearson's correlation coefficient (Pearson's r) between "Unit Price" and "Quantity"
correlation_coefficient = df['Unit price'].corr(df['Quantity'])
# Interpret the correlation
if correlation_coefficient > 0:
    correlation_direction = "positive"
elif correlation_coefficient < 0:
    correlation_direction = "negative"
else:
    correlation_direction = "no"
print(f"Pearson's Correlation Coefficient (r): {correlation_coefficient:.2f}")
print(f"There is a {correlation_direction} correlation between Unit Price and Quantity.")
```

### chi-squared (χ²) test between branch and payment variables

```python
import pandas as pd
from scipy.stats import chi2_contingency
# contingency table between "Branch" and "Payment"
contingency_table = pd.crosstab(df['Branch'], df['Payment'])
# Perform a chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)
# Define significance level (alpha)
alpha = 0.05
# Compare p-value to alpha to determine significance
if p < alpha:
    print("Null hypothesis rejected: There is a significant association between Branch and Payment.")
else:
    print("Null hypothesis not rejected: There is no significant association between Branch and Payment.")

print(f"Chi-squared statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
```

### z test of customer type 

```python
import pandas as pd
from scipy import stats
# Extract data for "Member" and "Normal" customer types
member_data = df[df['Customer type'] == 'Member']['Total']
normal_data = df[df['Customer type'] == 'Normal']['Total']
#  Z-test
z_stat, p_value = stats.ttest_ind(member_data, normal_data)
# (alpha)
alpha = 0.05
# Compare p-value to alpha to determine significance
if p_value < alpha:
    print("Null hypothesis rejected: There is a significant difference between Member and Normal customers.")
else:
    print("Null hypothesis not rejected: There is no significant difference between Member and Normal customers.")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```
